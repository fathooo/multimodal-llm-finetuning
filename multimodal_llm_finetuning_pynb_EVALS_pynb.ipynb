```python
#cargar el modelo aquí
from transformers import Trainer, TrainingArguments, ChameleonProcessor, ChameleonForConditionalGeneration, BitsAndBytesConfig
from peft import prepare_model_for_kbit_training
import torch

MODEL_NAME = "facebook/chameleon-7b"
USE_PEFT = True
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

quantization_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.bfloat16,
)

model = ChameleonForConditionalGeneration.from_pretrained(MODEL_NAME, quantization_config=quantization_config, device_map={"": 0})

if USE_PEFT:
    print("Preparando el modelo para el entrenamiento con PEFT (k-bit)")
    model.gradient_checkpointing_enable()
    model = prepare_model_for_kbit_training(model)

print(f"Modelo cargado en el dispositivo")
```

```python
processor = ChameleonProcessor.from_pretrained(MODEL_NAME)

# Prompt inicial que actúa como las instrucciones del 'system'
initial_instruction = "Responde únicamente con la letra correspondiente (A, B, C, D)."
question = "Cual es la capital de francia?"
options = "A) Madrid\nB) Paris\nC) Berlin\nD) Roma"

input_text = f"SYSTEM:{initial_instruction}\n\nPregunta: {question}\nOpciones: {options}\nRespuesta:"


inputs = processor(
    text=input_text,
    return_tensors="pt"
).to(device, dtype=torch.bfloat16)


outputs = model.generate(
    **inputs,
    max_new_tokens=1,
    num_beams=5,
    early_stopping=True,
    pad_token_id=processor.tokenizer.eos_token_id,
)

answer = processor.decode(outputs[0], skip_special_tokens=True)

if "Respuesta:" in answer:
    answer = answer.split("Respuesta:")[-1].strip()


print(f"Pregunta: {question}")
print(f"Opciones:\n{options}")
print(f"Respuesta generada: {answer}")
```
